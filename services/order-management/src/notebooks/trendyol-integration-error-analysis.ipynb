```python
// Update to normalizeOrders method in TrendyolService

async normalizeOrders(trendyolOrders) {
  try {
    const normalizedOrders = [];
    const skippedOrders = [];
    const defaultUserId = process.env.DEFAULT_USER_ID || '1';
    
    logger.debug(`Normalizing ${trendyolOrders.length} Trendyol orders with connection ID: ${this.connectionId}`);
    
    // Track counts for reporting
    let successCount = 0;
    let skippedCount = 0;
    
    for (const order of trendyolOrders) {
      try {
        // Check if order already exists by externalOrderId (orderNumber) AND platformId
        const existingOrder = await Order.findOne({
          where: {
            externalOrderId: order.orderNumber,
            platformId: this.connectionId
          }
        });
        
        if (existingOrder) {
          // Order already exists, skip it
          skippedOrders.push({
            externalOrderId: order.orderNumber,
            reason: 'Order already exists'
          });
          skippedCount++;
          continue;
        }
        
        // Get phone number from shipment address
        const phoneNumber = this.extractPhoneNumber(order);
        
        // First create the order record - ensure platformId is properly formatted
        const normalizedOrder = await Order.create({
          externalOrderId: order.orderNumber,
          platformId: String(this.connectionId), // Ensure it's a string
          userId: defaultUserId,
          connectionId: this.connectionId,
          orderDate: new Date(order.orderDate),
          status: this.mapOrderStatus(order.status),
          totalAmount: order.totalPrice,
          currency: 'TRY',
          customerName: `${order.customerFirstName} ${order.customerLastName}`,
          customerEmail: order.customerEmail,
          customerPhone: phoneNumber,
          notes: order.note || '',
          paymentStatus: 'pending',
          rawData: JSON.stringify(order)
        });
        
        // Rest of the code for creating shipping details and order items
        // ...
        
        normalizedOrders.push(normalizedOrder);
        successCount++;
      } catch (orderError) {
        // Handle individual order errors
        logger.error(`Failed to normalize individual Trendyol order: ${orderError.message}`, { 
          error: orderError, 
          orderNumber: order.orderNumber,
          connectionId: this.connectionId 
        });
        
        skippedOrders.push({
          externalOrderId: order.orderNumber,
          reason: orderError.message
        });
        skippedCount++;
      }
    }
    
    // Return both successful and skipped orders
    return {
      orders: normalizedOrders,
      skipped: skippedOrders,
      stats: {
        success: successCount,
        skipped: skippedCount,
        total: trendyolOrders.length
      }
    };
  } catch (error) {
    logger.error(`Failed to normalize Trendyol orders: ${error.message}`, { error, connectionId: this.connectionId });
    throw error;
  }
}
```
